```python
col_1_mapping = {
    'A': 1, # rock
    'B': 2, # paper
    'C': 3 # scissors
}
```

```python
col_2_intial_mapping = {
    'X': 1, # rock
    'Y': 2, # paper
    'Z': 3 # scissors
}
```

```python
round_scores = {
    1: 6, # win
    0: 3, # draw
    -1: 0 # loss
}
round_scores_inverted = {value: key for key, value in round_scores.items()}
```

```python
def calculate_round1_score(col_1_move, col_2_move):
  move_1_score = col_1_mapping[col_1_move]
  move_2_score = col_2_intial_mapping[col_2_move]
  round_outcome = (move_2_score - move_1_score) % 3

  score = round_scores.get(round_outcome, 0)
  # print(round_outcome, score, move_2_score)
  return score + move_2_score    
```

```python
col_2_final_mapping = {
    'X': 0, # lose
    'Y': 3,  # draw
    'Z': 6 # win
}
  
def calculate_round2_score(col_1_move, col_2_outcome):
  move_1_score = col_1_mapping[col_1_move]
  score = col_2_final_mapping.get(col_2_outcome, 0)

  move_2_score = (move_1_score + round_scores_inverted.get(score, 0)) % 3
  
  if move_2_score == 0:
    move_2_score = 3

  # print(col_1_move, col_2_outcome, move_1_score, move_2_score, score)

  return move_2_score + score
```

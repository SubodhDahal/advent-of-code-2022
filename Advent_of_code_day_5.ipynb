```python
# crate_configs = crate_configs_example
# instructions = instructions_example
crate_configs = crate_configs_task
instructions = instructions_task
```

```python
def get_split_crates (crate_configs):
  split = []
  for line in crate_configs.split('\n'):
    split.append(line)
  split.remove('')
  split.remove('')
  return split
split_crates = get_split_crates(crate_configs)
split_crates
```

```python
crates = split_crates[-1]
crates
```

```python
crates_count = len(crates.replace(' ', ''))
crates_count
```

```python
import re

regex = re.compile(r"move (\d+) from (\d) to (\d)")
```

```python
def get_machine_instructions(instructions):
  movements = []
  for line in instructions.split('\n'):
    match = regex.match(line)
    quantity, origin, destination = match.groups()
    movements.append((quantity, origin, destination))
  return movements  
movement_instructions = get_machine_instructions(instructions)
movement_instructions
```

```python
def get_crate_contents (crate_number):
  crate_contents = []
  crate_position = crates.find(crate_number)
  # print(crate_position)
  for line_content in split_crates[:-1]:
    crate_content = line_content[crate_position]
    if crate_content != ' ':
      crate_contents.append(crate_content)
  crate_contents.reverse()
  return crate_contents

def get_all_crate_contents ():
  crates_count = len(crates.replace(' ', ''))
  return [get_crate_contents(str(number)) for number in range(1, crates_count + 1)]

crate_contents = get_all_crate_contents()
crate_contents
```

## Part 1

```python
def get_final_crate_configuration_strategy_1 (crate_contents):
  copied_crate = crate_contents[:]
  for (quantity, origin, destination) in movement_instructions:
    for i in range(int(quantity)):
      if len(copied_crate[int(origin) - 1]) > 0:
        change = copied_crate[int(origin) - 1].pop()
        copied_crate[int(destination) - 1].append(change)
  return copied_crate  

final_crate_config = get_final_crate_configuration_strategy_1(get_all_crate_contents())
''.join([crate[-1] for crate in final_crate_config])
```

## Part 2

```python
def get_final_crate_configuration_strategy_2 (crate_contents):
  copied_crate = crate_contents[:]
  for (quantity, origin, destination) in movement_instructions:
    changes = []
    for i in range(int(quantity)):
      if len(copied_crate[int(origin) - 1]) > 0:
        change = copied_crate[int(origin) - 1].pop()
        changes.append(change)
    changes.reverse()
    for change in changes:
      copied_crate[int(destination) - 1].append(change)
  return copied_crate  

final_crate_config = get_final_crate_configuration_strategy_2(get_all_crate_contents())
''.join([crate[-1] for crate in final_crate_config])
```
